```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

**Loading the Data**

```python
df = pd.read_csv("Iris.csv")
df.head()
```

**Getting to know the dataset**

```python
df.Species.value_counts()
```

```python
df.drop(["Id"], axis = 1, inplace=True)
```

```python
df.head()
```

```python
df["Species"] = df["Species"].str.replace("Iris-","")
```

```python
df.head()
```

```python
df.info()
```

**EDA**

```python
sns.pairplot(df, hue = "Species", height = 2.75)
```

```python
plt.show()
```

```python
df["Species_Num"] = df["Species"]
```

```python
df.head()
```

```python
df["Species_Num"] = df["Species_Num"].str.replace("setosa","1")
df["Species_Num"] = df["Species_Num"].str.replace("versicolor","2")
df["Species_Num"] = df["Species_Num"].str.replace("virginica","3")
```

```python
df.head()
```

```python
df.info()
```

```python
df["Species_Num"] = pd.to_numeric(df["Species_Num"], downcast="integer")
```

```python
df.head()
```

```python
df.info()
```

```python
sns.heatmap(df.select_dtypes(include=['number']).corr(), 
            cmap="Spectral", annot=True)
```

```python
plt.figure(figsize = (15, 9))
sns.heatmap(df.select_dtypes(include=['number']).corr(), 
            cmap="Spectral", annot=True)
plt.show()
```

```python
df.drop(["PetalLengthCm", "Species_Num"], axis=1, inplace=True)
```

```python
df.head()
```

**Decision Tree Modelling**

```python
X = df.drop(["Species"], axis = 1)
y = df.Species
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier()
```

```python
dt_model.fit(X_train, y_train)
```

```python
y_pred = dt_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print("Accuracy Score:", accuracy_score(y_test, y_pred))
```
